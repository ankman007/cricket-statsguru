Clean the csv files here!

Steps to clean the data: 
- First change the '-' into Null.
- We either can change the data type from strings to integer now or we can do it while visualizing in streamlit.

```python
#importing library 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
batting_players_odi = pd.read_csv('data/batting_players_odi.csv')
```

```python
batting_players_odi.info()
```

```python
#changin the "-" to null.
for i in range(batting_players_odi.shape[0]):
    for j in range(batting_players_odi.shape[1]):
        if batting_players_odi.iloc[i , j] == "-":
            batting_players_odi.iloc[i , j] = None
```

```python
np.array(batting_players_odi['HS'].str.extract('(\d+)').astype(float)).reshape(-1) #we will use this to extract the required columns as float
```

```python
#Using this function to clean the dataframe
def clean_dataframe(df: pd.DataFrame):
    for i in range(df.shape[0]):
        for j in range(df.shape[1]):
            if df.iloc[i , j] == "-":
                df.iloc[i , j] = None
```

```python
bowling_players_odi = pd.read_csv('data/bowling_players_odi.csv')
clean_dataframe(bowling_players_odi)
```
